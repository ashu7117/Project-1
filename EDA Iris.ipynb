```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv("iris.csv")
```

```python
# Display the first few rows of the dataset
print(iris.head())
```

```python
print(iris.shape)
```

```python
print(iris.columns)
```

```python
iris["variety"].value_counts()
```

```python
# Histograms for each feature
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
sns.histplot(iris['sepal.length'], bins=20, kde=True, color='blue')
plt.title('Sepal Length Distribution')
```

```python
# Histograms for each feature
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
sns.histplot(iris['petal.width'], bins=20, kde=True, color='blue')
plt.title('Sepal Length Distribution')

# Similar histograms for sepal_width, petal_length, and petal_width
```

```python
# Pairplot for visualizing relationships between features
sns.pairplot(iris, hue="variety", markers=["o", "s", "D"])
```

```python
# Box plots to visualize the distribution and identify outliers
plt.figure(figsize=(14, 8))
plt.subplot(2, 2, 1)
sns.boxplot(x='variety', y='sepal.length', data=iris)
```

```python
# Box plots to visualize the distribution and identify outliers
plt.figure(figsize=(14, 8))
plt.subplot(2, 2, 1)
sns.boxplot(x='variety', y='petal.width', data=iris)
```

## 2-D Scatter Plot

```python
iris.plot(kind='scatter', x='sepal.length', y='sepal.width');
plt.show()
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(iris, hue="variety", size=4)  \
   .map(plt.scatter, "sepal.length", "sepal.width")  \
   .add_legend()
plt.show()
```

```python
# Observations
1. Using sepal.length and sepal.width features, we can distinguish Setosa flowers from others.
2. Separating Versicolor from Viginica is much harder as they have considerable overlap. 
```
